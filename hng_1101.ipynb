## Identify Neighbouring Polling Units
We define neighbours based on a radius (e.g., 1 km) and calculate the geodesic distance between each polling unit.

```python
print(new_df.columns)
```

```python
# Remove rows with NaN values in 'latitude' or 'longitude'
clean_df = new_df.dropna(subset=['Latitude', 'Longitude'])

# Ensure there are no NaN values in latitude and longitude
clean_df['Latitude'] = pd.to_numeric(clean_df['Latitude'], errors='coerce')
clean_df['Longitude'] = pd.to_numeric(clean_df['Longitude'], errors='coerce')

# Drop any rows where conversion to numeric failed
clean_df = clean_df.dropna(subset=['Latitude', 'Longitude'])

# Final check
print(clean_df[['Latitude', 'Longitude']].isna().sum())
print(clean_df[['Latitude', 'Longitude']].describe())
```

```python
from geopy.distance import geodesic

# Function to find neighbours within a radius
def find_neighbours(lat, lon, clean_df, radius=1):
    neighbours = []
    for index, row in clean_df.iterrows():
        if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
            dist = geodesic((lat, lon), (row['Latitude'], row['Longitude'])).km
            if dist <= radius:
                neighbours.append(index)
    return neighbours

# Apply the function to find neighbours
clean_df['neighbours'] = clean_df.apply(lambda row: find_neighbours(row['Latitude'], row['Longitude'], clean_df), axis=1)
```

This step finds neighbouring polling units within a defined radius, which is crucial for comparing votes.

## Calculate Outlier Scores
We calculate the outlier scores by comparing votes for each party with the votes of neighbouring units.

```python
# Function to calculate outlier score
def calculate_outlier_score(votes, neighbours_votes):
    if len(neighbours_votes) > 0:
        return abs(votes - neighbours_votes.mean())
    else:
        return None

# Calculate outlier scores for each party
parties = ['APC', 'LP', 'PDP', 'NNPP']
for party in parties:
    clean_df[f'{party}_outlier_score'] = clean_df.apply(lambda row: calculate_outlier_score(row[party], clean_df.loc[row['neighbours'], party]), axis=1)

# Save the dataset with outlier scores
clean_df.to_csv('/home/tina/Self_Projects/DELTA_crosschecked_with_outlier_scores.csv', index=False)
```

This step computes the outlier scores, which help identify polling units with significant vote deviations from their neighbours.

```python
import folium

# Create a base map
m = folium.Map(location=[clean_df['Latitude'].mean(), clean_df['Longitude'].mean()], zoom_start=10)

# Add polling units to the map
for i, row in clean_df.iterrows():
    folium.CircleMarker(
        location=(row['Latitude'], row['Longitude']),
        radius=5,
        popup=f"PU-Code': {row['PU-Code']}\nAPC_outlier_score: {row['APC_outlier_score']}\nLP_outlier_score: {row['LP_outlier_score']}\nPDP_outlier_score: {row['PDP_outlier_score']}\nNNPP_outlier_score: {row['NNPP_outlier_score']}",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(m)

# Highlight top 3 outliers in red
top_outliers = clean_df.nlargest(3, [f'{party}_outlier_score' for party in ['APC', 'LP', 'PDP', 'NNPP']])
for i, row in top_outliers.iterrows():
    folium.CircleMarker(
        location=(row['Latitude'], row['Longitude']),
        radius=8,
        popup=f"PU-Code': {row['PU-Code']}\nAPC_outlier_score: {row['APC_outlier_score']}\nLP_outlier_score: {row['LP_outlier_score']}\nPDP_outlier_score: {row['PDP_outlier_score']}\nNNPP_outlier_score: {row['NNPP_outlier_score']}",
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(m)

# Save the map
m.save('/home/tina/Self_Projects/DELTA_crosschecked_polling_units_map.html')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Create a boxplot for outlier scores of each party
plt.figure(figsize=(10, 6))
sns.boxplot(data=clean_df[[f'{party}_outlier_score' for party in ['APC', 'LP', 'PDP', 'NNPP']]])
plt.title('Distribution of Outlier Scores by Party')
plt.xlabel('Party')
plt.ylabel('Outlier Score')
plt.savefig('/home/tina/Self_Projects/DELTA_crosschecked_with_outlier_scores_boxplot.png')
plt.show()

# Create histograms for outlier scores of each party
plt.figure(figsize=(12, 8))
for i, party in enumerate(['APC', 'LP', 'PDP', 'NNPP']):
    plt.subplot(2, 2, i+1)
    sns.histplot(clean_df[f'{party}_outlier_score'], bins=30, kde=True)
    plt.title(f'Distribution of {party} Outlier Scores')
    plt.xlabel('Outlier Score')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('/home/tina/Self_Projects/DELTA_crosschecked_with_outlier_scores_histograms.png')
plt.show()
```
